```python
import json 
import pandas as pd 
import numpy as np 
import pylab as plt 
import networkx as nx
```

```python
def loadjson(fname):
    with open(fname, 'r') as fdata:
        res = json.load(fdata)

    res = {key: val['order'] for key, val in res.items()}
    count = 0 
    for val in res.values():
        if val[-1] == 'random' and "sheppard" not in fname:
            count += 1
        if val[0] == 'random' and "sheppard" in fname:
            count += 1
    res = {"random-last": count / len(res.values())} | res
    return res
```

```python
resnames = ["normalized-stress","minopt", "sheppard", "sheppardscale", "raw", "kruskal", "unitball"]
dicts = dict(zip(
    resnames,
    [loadjson(f"results/{s}-results.json") for s in resnames]
))
df = pd.DataFrame(dicts)
df.head()
```

```python
df['sheppard'] = df["sheppard"].apply(lambda x: list(reversed(x)) if isinstance(x, list) else x)
```

```python
df.to_csv("out-results.csv")
```
