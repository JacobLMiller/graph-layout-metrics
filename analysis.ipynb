```python
import json 
import pandas as pd
```

```python
def loadjson(fname):
    with open(fname, 'r') as fdata:
        res = json.load(fdata)

    res = {key: val['order'] for key, val in res.items()}
    
    return res
```

```python
# Create dataframe from graphs in "results/*-results.json"

resnames = ["raw", "kk", "normalized-stress", "minopt", "sheppard", "sheppardscale", "ratios", "kruskal"]
# resnames = ["raw", "kk", "normalized-stress", "minopt", "sheppard", "sheppardscale", "kruskal"]

dicts = dict(zip(
    resnames,
    [loadjson(f"results/{s}-results.json") for s in resnames]
))
df = pd.DataFrame(dicts)
```

```python
# Shepard Goodness Score needs to be reversed
df['sheppard'] = df['sheppard'].apply(lambda x: list(reversed(x)) if isinstance(x, list) else x)
```

```python
# Save file
df.to_csv("results/out-results.csv")
df.head()
```
